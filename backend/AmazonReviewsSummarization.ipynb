```python
import pandas as pd
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
def spacy_summarizer(text):
    doc = nlp(text)
    summary = ' '.join([chunk.text for chunk in doc.noun_chunks])
    return summary
```

```python
reviews_df['clean_review_text'] = reviews_df['clean_review_text'].fillna('').astype(str)

reviews_df['spacy_summary'] = reviews_df['clean_review_text'].apply(spacy_summarizer)
```

```python
grouped = reviews_df.groupby('product').agg(
    total_reviews=('review', 'count'),
    avg_rating=('rating', 'mean'),
    summary = ('clean_review_text', lambda x: ' '.join(x))
).reset_index()
```

```python
grouped
```

```python
grouped['summary'] = grouped['summary'].fillna('').astype(str)

grouped['spacy_summary'] = grouped['summary'].apply(spacy_summarizer)
```
